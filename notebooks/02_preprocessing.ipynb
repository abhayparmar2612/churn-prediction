```python
import pandas as pd
```

```python
df = pd.read_csv('clean_churn_data.csv')
```

```python
x =df.drop('Churn',axis=1)
```

```python
y = df['Churn']
```

```python
cat_cols=x.select_dtypes(include='object').columns.tolist()
```

```python
cat_cols
```

```python
num_cols=x.select_dtypes(include=['int64','float64']).columns.tolist()
```

```python
num_cols
```

```python
import sklearn 
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test= train_test_split(x,y,test_size=0.2, stratify=y,random_state=42)
```

### splitting is done 

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
preprocessor=ColumnTransformer(
    transformers= [
        ('num',StandardScaler(),num_cols),
        ('cat',OneHotEncoder(handle_unknown='ignore'),cat_cols)
    ]
)
```

```python
preprocessor
```

```python
X_train_transformed = preprocessor.fit_transform(X_train)
X_test_transformed = preprocessor.transform(X_test)
```

```python
print("Transformed X_train shape:", X_train_transformed.shape)
```

```python
import joblib
```

```python
joblib.dump(preprocessor,'preprocessor.pkl')
```

### Preprocessing Summary

- Applied StandardScaler to numerical features
- Applied OneHotEncoder to categorical features
- Used ColumnTransformer to create clean, modular preprocessing
- Saved the preprocessor to use during model training and deployment
- Train/Test split maintained class balance with stratification
